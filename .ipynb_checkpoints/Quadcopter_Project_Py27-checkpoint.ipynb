## Running the two cells below works fine with my tf-gpu kernel build with Python 3.5         as I show in my other notebook, but here with Python2.5 I get an error

```python
#  trying to set seeds
from numpy.random import seed
seed(3)
from tensorflow import set_random_seed
set_random_seed(3)
```

```python
## this ran fine with tf-gpu and python3.5
## but I had to move code down from agents to base to work with tf-gpu2 and python2.7
#!/usr.bin/env python2.7
import sys
import pandas as pd
import numpy as np
from agent import DDPG
from takeoff import Task
import csv
# done above   np.random.seed(0)    # so I can reproduce things


num_episodes = 100  # note: changed this from 1000 to 10 for starters
target_pos = np.array([0., 0., 100.])
init_pos = np.array([0., 0., 0., 0., 0., 0.])
task = Task(init_pose=init_pos, target_pos=target_pos)
agent = DDPG(task) 
best_score = -1000.
file_output = "July14rewards.txt"
rewards_labels = ['Episode', 'Score']
rewards_data = {x: [] for x in rewards_labels}
with open(file_output, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(rewards_labels)

    for i_episode in range(1, num_episodes+1):
        state = agent.reset_episode() # start a new episode
        score = 0
        while True:
            action = agent.act(state) 
            #print("state is {}".format(state))
            #print(action)
            next_state, reward, done = task.step(action)
            #print("after taking action, we got reward of {:7.3f}".format(reward))
            agent.step(action, reward, next_state, done)
            state = next_state
            score += reward
           
            best_score = max(best_score, score)
            if done: 
                to_write = [i_episode, score]
                for ii in range(len(rewards_labels)):
                    rewards_data[rewards_labels[ii]].append(to_write[ii])
                writer.writerow(to_write)
                if i_episode % 10 == 0 or i_episode <=3:
                    print("\rEpisode = {:4d}, score = {:7.3f} (best = {:7.3f})".format(
                        i_episode, score, best_score, end=""))  # [debug]
                #print(action) 
                #print(next_state)
                break
        rewards_data['Episode'].append(i_episode)    
        rewards_data['Score'].append(score)
sys.stdout.flush()
    
```
